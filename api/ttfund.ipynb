```python
import requests
```

```python
API = 'http://fund.eastmoney.com/api/FundGuide.aspx?'
DEFAULT_PARAMS = {
        'dt':0,
        'ft':'',
        'sd':'',
        'ed':'', 
        'sc':'z',
        'st':'desc',
        'pi':1,
        'pn':20,
        'zf':'diy',
        'sh':'list',
        'rnd':0.3874453763350052
}


class FundSearch(object):
    '''
    Api for tiantian fund filter
    '''
    def __init__(self,):
        self.api_name = 'FundGuide'
        
    def info(self, **info):
        return self.api_name+':'+info.__str__()

    def search(self,**params):
        get_params = '&'.join([f'{k}={v}' for k,v in DEFAULT_PARAMS.items()]+\
        [f'{k}={v}' for k,v in params.items()])
        response = requests.get(API+get_params)
        return response
```

```python
fs = FundSearch()
```

```python
fs.api_name
```

```python
fs.info(a='b',c='d')
```

```python
res = fs.search()
```
